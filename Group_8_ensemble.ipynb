```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import VotingClassifier, StackingClassifier, BaggingClassifier, RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingRegressor
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import NuSVC
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
df=pd.read_csv('Group_8_data_cleaned.csv')#reading csv file using pandas
df.head()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

X = df[[ "air_temp","process_temp","rot_speed","torque","tool_wear","Type_H","Type_L","Type_M"]]
y = df["failure"]
```

```python
from sklearn.model_selection import train_test_split #splitting the data into training ,validation and test data
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)#splitting data into training and testing 

X_train1, X_valid, y_train1, y_valid = train_test_split(X_train_full, y_train_full, stratify=y_train_full, test_size=0.25, random_state=42)#splitting training data into traiing and validation data
```

```python
from imblearn.over_sampling import SMOTE
X_train, y_train = SMOTE().fit_resample(X_train1,y_train1)
```

```python
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)

X_test_scaled = scaler.transform(X_test)
X_valid_scaled = scaler.transform(X_valid)
```

```python
clf1=LogisticRegression(penalty= 'l2', class_weight= 'balanced')
clf2=KNeighborsClassifier(weights= 'distance', p= 2, n_neighbors= 1)
clf3=RandomForestClassifier(n_estimators= 46, min_samples_split= 2, min_samples_leaf= 2, max_features='sqrt',
                                             max_depth= 18)
clf4=DecisionTreeClassifier(max_depth= 19, max_features= 'auto',min_samples_leaf= 2, min_samples_split= 2)
clf5=SVC(kernel= 'linear', C= 0.01)
clf6=LinearSVC(C= 0.01)
clf7=SVC(kernel= 'poly', gamma= 0.1, degree= 2, C =5.0)
clf8=SVC(kernel= 'rbf', gamma= 1, C= 100)
clf9=SVC(C= 100, gamma= 0.1, kernel= 'rbf')
clf10=DecisionTreeClassifier(random_state=1)
clf11=SGDClassifier(penalty='l2', loss= 'hinge', learning_rate= 'adaptive', eta0= 100, class_weight= {1: 0.6, 0: 0.4}, alpha=0.0001)
clf12=RandomForestClassifier()
clf13=RandomForestClassifier(max_depth=70, max_features='log2', n_estimators=62)
clf14=AdaBoostClassifier()
clf15=GradientBoostingClassifier()
clf16=LinearDiscriminantAnalysis()
clf17=NuSVC(probability=True)
clf18=QuadraticDiscriminantAnalysis()
clf19=GaussianNB()
clf20=LogisticRegression()
VC = VotingClassifier(
    estimators=[('LR', clf1),
                                  ('KNN', clf2),
                                  ('RF1', clf3),
                                 ('DT1',clf4),
                                 ('SVC_LINEAR1',clf5),
                                  ('SVC_LINEAR2',clf6),
                                  ('SVC_POLY',clf7),('SVC_RBF1',clf8),('SVC_RBF2',clf9),
                                ('DT2', clf10),
                                  ('SGD', clf11),
                                  ('RF2', clf12),
                                 ('RF3',clf13),
                                 ('ADA_BOOST',clf14),
                                  ('GRAD_BOOST',clf15),
                                  ('LIN_DISCRIM',clf16),('NuSVC',clf17),('Quad_Dis',clf18),('Gauss_NB',clf19),('LR1',clf20) ], 
                      voting='hard')

for clf, name in zip([clf1, clf2, clf3,clf4,clf5,clf6,clf7,clf8,clf9,clf10,clf11, clf12, clf13,clf14,clf15,clf16,clf17,clf18,clf19,clf20],['LR','KNN','RF1','DT1','SVC_LINEAR1','SVC_LINEAR2','SVC_POLY','SVC_RBF1','SVC_RBF2','DT2','SGD','RF2','RF3','ADA_BOOST','GRAD_BOOST','LIN_DISCRIM','NuSVC','Quad_Dis','Gauss_NB',"LR1"]):
    clf.fit(X_train_scaled, y_train)
    print(name,'Training accuracy:', clf.score(X_train_scaled, y_train))
    print(name, 'Test accuracy:', clf.score(X_valid_scaled, y_valid), '\n')
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
VC.fit(X_train, y_train)
forest_y_pred = VC.predict(X_valid)
cm=confusion_matrix(y_valid, forest_y_pred)
print(cm)
print(classification_report(y_valid, forest_y_pred), "\n")
```
